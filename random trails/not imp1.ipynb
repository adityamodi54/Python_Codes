```python
import urllib.request
import lxml.html
connection = urllib.request.urlopen('http://py4e-data.dr-chuck.net/known_by_Fikret.html')

dom =  lxml.html.fromstring(connection.read())

for link in dom.xpath('//a/@href'): # select the url in href for all a tags(links)
    print (link)
```

```python
import urllib.request
from bs4 import BeautifulSoup
url = urllib.request.urlopen('http://py4e-data.dr-chuck.net/known_by_Fikret.html').read()
soup = BeautifulSoup(url)
for line in soup.find_all('a'):
    print(line.get('href'))
```

```python
from bs4 import BeautifulSoup
import urllib.request
import re
 
html_page = urllib.request.urlopen("http://arstechnica.com")
soup = BeautifulSoup(html_page)
links = []
 
for link in soup.findAll('a', attrs={'href': re.compile("^http://")}):
    links.append(link.get('href'))
 
print(links)
```
